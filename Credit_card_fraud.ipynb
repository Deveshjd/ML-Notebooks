```python
x = df.drop('Class', axis = 1)
x.head()
```

```python
y = df['Class']
```

```python
corr_matrix = df.corr()

plt.figure(figsize = (20,15))
sns.heatmap(corr_matrix, annot = True, fmt = '.1f', cmap = 'YlGnBu', vmin=-1, vmax=1)
plt.show()
```

```python
# From the heatmap we can see that the features V2 and V5 have very strong negative correlation with Amount.
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# If you want to train different classification models:
# Just uncomment the model name and execute the cells again.

models_dict = {
    "LogReg" : LogisticRegression(),
    # "DTC" : DecisionTreeClassifier(),
    # "RMC" : RandomForestClassifier(),
    # "SVC" : SVC()
}
for name,model in models_dict.items():
  model.fit(x_train_scaled, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score
```

```python
for name,model in models_dict.items():
    y_pred = model.predict(x_test_scaled)

    print(f"==={name}===")
    print(f"Accuracy : {accuracy_score(y_test, y_pred)}")
    print(f"Precision : {precision_score(y_test, y_pred)}")
    print(f"Recall : {recall_score(y_test, y_pred)}")
    print(f"f1 score : {f1_score(y_test, y_pred)}")

    print("Confusion Matrix")
    sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, fmt = '.2f')
    plt.show()
```

*From the above result of logistic regression we can see that the model's accuracy is excellent but the recall is not appreciable. It happenned due to imbalanced dataset. Earlier we saw that there is huge difference between the no. of valid and fraud data in the dataset*

*If we want to increase the model performance then we have do data preprocessing and try to make the data balanced.*
